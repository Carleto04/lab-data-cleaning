#### 4. Import the users table.

```python
users_table = pd.read_csv("../users.csv", )
users_table = users_table.drop("Unnamed: 0", axis = 1) #drop index unnecessary
users_table
```

#### 5. Rename Id column to userId.

```python
users_table.rename(columns = {'userId':'Id'},inplace = True) #change column name
users_table.rename(columns = {'Id':'userId'},inplace = True)
users_table
```

#### 6. Import the posts table. 

```python
posts_table = pd.read_csv("../posts.csv")
posts_table = posts_table.drop("Unnamed: 0", axis = 1) #drop index unnecessary
posts_table
```

#### 7. Rename Id column to postId and OwnerUserId to userId.

```python
posts_table.rename(columns = {'PostId':'Id'},inplace = True)
posts_table.rename(columns = {'Id':'PostId'},inplace = True)
posts_table
```

#### 8. Define new dataframes for users and posts with the following selected columns:
**users_sliced columns**: userId, Reputation, Views, UpVotes, DownVotes  
**posts_sliced columns**: postId, Score, userId, ViewCount, CommentCount

```python
df_users = users_table[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']] #change columns name
df_posts = posts_table[['PostId', 'Score', 'userId', 'ViewCount', 'CommentCount']]
df_users
```

#### 9. Merge the two dataframes created in the step above (8), users_sliced and posts_sliced. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merged_df = pd.merge(left=df_posts, right=df_users, how= 'inner')
merged_df
```
